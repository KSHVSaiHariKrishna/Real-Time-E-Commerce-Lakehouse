```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window
```

```python
# ── Silver base tables ─────────────────────────────────────────────────────
silver_orders      = spark.table("ecommerce.silver.orders")
silver_products    = spark.table("ecommerce.silver.products")
silver_customers   = spark.table("ecommerce.silver.customers")
silver_inventory   = spark.table("ecommerce.silver.inventory")
silver_clickstream = spark.table("ecommerce.silver.clickstream")

# ── Silver mart tables ─────────────────────────────────────────────────────
orders_clickstream = spark.table("ecommerce.silver.orders_clickstream")
inventory_orders   = spark.table("ecommerce.silver.inventory_orders")
customer_360       = spark.table("ecommerce.silver.customer_360")

print("✅ All Silver tables loaded")
```

```python

# ─────────────────────────────────────────────────────────────────────────────
# 1. revenue_daily
# ─────────────────────────────────────────────────────────────────────────────
revenue_daily = (
    silver_orders
    .filter(col("is_delivered") == True)
    .groupBy(
        "order_date_only",
        "order_year",
        "order_month",
        "order_quarter",
        "order_day_name",
        "is_weekend",
    )
    .agg(
        count("order_id").alias("total_orders"),
        sum("total_amount").alias("total_revenue"),
        round(avg("total_amount"), 2).alias("avg_order_value"),
        sum("quantity").alias("total_items_sold"),
        countDistinct("customer_id").alias("unique_customers"),
        countDistinct("product_id").alias("unique_products_sold"),
    )
    .withColumn("revenue_per_customer",
        round(col("total_revenue") / col("unique_customers"), 2))
    .withColumn("silver_updated_at", current_timestamp())
    .orderBy("order_date_only")
)

(
    revenue_daily.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .option("path", "s3://ecommerce-lakehouse-databricks/gold/revenue_daily/")
    .saveAsTable("ecommerce.gold.revenue_daily")
)

print(f"✅ gold.revenue_daily → {revenue_daily.count():,} rows")
revenue_daily.show(10)
```

```python

# ─────────────────────────────────────────────────────────────────────────────
# 2. customer_segments
# ─────────────────────────────────────────────────────────────────────────────
customer_segments = (
    customer_360
    .groupBy(
        "customer_value_segment",
        "loyalty_tag",
        "preferred_device",
        "is_active",
    )
    .agg(
        count("customer_id").alias("customer_count"),
        round(avg("total_spend"),         2).alias("avg_total_spend"),
        round(avg("total_orders"),        2).alias("avg_orders"),
        round(avg("avg_order_value"),     2).alias("avg_order_value"),
        round(avg("cart_conversion_rate"), 2).alias("avg_cart_conversion_rate"),
        round(avg("total_sessions"),      2).alias("avg_sessions"),
        round(avg("abandoned_carts"),     2).alias("avg_abandoned_carts"),
        round(sum("total_spend"),         2).alias("segment_total_revenue"),
    )
    .withColumn("segment_revenue_pct",
        round(
            col("segment_total_revenue")
            / sum("segment_total_revenue").over(Window.partitionBy())
            * 100
        , 2))
    .withColumn("silver_updated_at", current_timestamp())
    .orderBy(desc("segment_total_revenue"))
)

(
    customer_segments.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .option("path", "s3://ecommerce-lakehouse-databricks/gold/customer_segments/")
    .saveAsTable("ecommerce.gold.customer_segments")
)

print(f"✅ gold.customer_segments → {customer_segments.count():,} rows")
customer_segments.show(10)

```

```python
# ─────────────────────────────────────────────────────────────────────────────
# 3. top_products
# ─────────────────────────────────────────────────────────────────────────────
top_products = (
    silver_orders
    .filter(col("is_delivered") == True)
    .groupBy(
        "product_id",
        "product_name",
        "category",
        "brand",
        "price_band",
        "order_year",
        "order_quarter",
    )
    .agg(
        count("order_id").alias("total_orders"),
        sum("quantity").alias("total_qty_sold"),
        round(sum("total_amount"), 2).alias("total_revenue"),
        round(avg("total_amount"), 2).alias("avg_order_value"),
        countDistinct("customer_id").alias("unique_buyers"),
        round(avg("margin_pct"), 2).alias("avg_margin_pct"),   # FIX: added round()
    )
    .withColumn("rank_in_category",
        rank().over(
            Window.partitionBy("category", "order_year", "order_quarter")
            .orderBy(desc("total_revenue"))
        ))
    .withColumn("rank_overall",
        rank().over(
            Window.partitionBy("order_year", "order_quarter")
            .orderBy(desc("total_revenue"))
        ))
    .withColumn("silver_updated_at", current_timestamp())
    .orderBy("order_year", "order_quarter", "rank_overall")
)

(
    top_products.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .option("path", "s3://ecommerce-lakehouse-databricks/gold/top_products/")
    .saveAsTable("ecommerce.gold.top_products")
)

print(f"✅ gold.top_products → {top_products.count():,} rows")
top_products.filter(col("rank_overall") <= 10).show(10)
```

```python

# ─────────────────────────────────────────────────────────────────────────────
# 4. restock_alerts
# ─────────────────────────────────────────────────────────────────────────────
# FIX: `warehouse_id` does not exist in inventory_orders.
#       Replaced with `warehouse_list` (array of warehouse IDs) and
#       `warehouse_count` (number of warehouses stocking the product).
restock_alerts = (
    inventory_orders
    .filter(
        (col("needs_restock") == True) |
        (col("demand_health").isin(
            "Critical — Restock Now",
            "Low Stock vs Demand"
        ))
    )
    .select(
        "product_id",
        "product_name",
        "category",
        "brand",
        "warehouse_list",          # FIX: was warehouse_id (not in schema)
        "warehouse_count",         # FIX: added — useful context alongside warehouse_list
        "available_qty",
        "stock_health",
        "demand_health",
        "total_qty_ordered",
        "qty_delivered",
        "total_revenue_generated",
        "stock_to_demand_ratio",
        "days_of_stock_remaining",
        "total_inventory_value",
        "price",
        "last_updated",
        "days_since_update",
    )
    .withColumn("restock_urgency",
        when(
            (col("available_qty") == 0) &
            (col("total_qty_ordered") > 0),    lit("CRITICAL — Out of Stock"))
        .when(col("days_of_stock_remaining") < 7,   lit("HIGH — < 7 days stock"))
        .when(col("days_of_stock_remaining") < 30,  lit("MEDIUM — < 30 days stock"))
        .otherwise(lit("LOW")))
    .withColumn("suggested_restock_qty",
        when(col("total_qty_ordered") > 0,
             round(col("total_qty_ordered") / 12 * 2, 0))
        .otherwise(lit(50)))
    .withColumn("silver_updated_at", current_timestamp())
    .orderBy("restock_urgency", desc("total_revenue_generated"))
)

(
    restock_alerts.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .option("path", "s3://ecommerce-lakehouse-databricks/gold/restock_alerts/")
    .saveAsTable("ecommerce.gold.restock_alerts")
)

print(f"✅ gold.restock_alerts → {restock_alerts.count():,} rows")
restock_alerts.show(10)
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# 5. funnel_analysis
# ─────────────────────────────────────────────────────────────────────────────
funnel_analysis = (
    silver_clickstream
    .groupBy("event_date", "event_year", "event_month")
    .agg(
        count("event_id").alias("total_events"),
        sum(when(col("is_page_view")    == True, lit(1)).otherwise(lit(0))).alias("page_views"),
        sum(when(col("is_product_view") == True, lit(1)).otherwise(lit(0))).alias("product_views"),
        sum(when(col("is_add_to_cart")  == True, lit(1)).otherwise(lit(0))).alias("add_to_carts"),
        sum(when(col("is_purchase")     == True, lit(1)).otherwise(lit(0))).alias("purchases"),
        countDistinct(
            when(col("is_page_view")    == True, col("user_id"))
        ).alias("users_page_view"),
        countDistinct(
            when(col("is_product_view") == True, col("user_id"))
        ).alias("users_product_view"),
        countDistinct(
            when(col("is_add_to_cart")  == True, col("user_id"))
        ).alias("users_add_to_cart"),
        countDistinct(
            when(col("is_purchase")     == True, col("user_id"))
        ).alias("users_purchased"),
        countDistinct("session_id").alias("total_sessions"),
    )
    .withColumn("page_to_product_rate",
        round(col("product_views") / col("page_views") * 100, 2))
    .withColumn("product_to_cart_rate",
        round(col("add_to_carts") / col("product_views") * 100, 2))
    .withColumn("cart_to_purchase_rate",
        round(col("purchases") / col("add_to_carts") * 100, 2))
    .withColumn("overall_conversion_rate",
        round(col("purchases") / col("page_views") * 100, 2))
    .withColumn("silver_updated_at", current_timestamp())
    .orderBy("event_date")
)

(
    funnel_analysis.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .option("path", "s3://ecommerce-lakehouse-databricks/gold/funnel_analysis/")
    .saveAsTable("ecommerce.gold.funnel_analysis")
)

print(f"✅ gold.funnel_analysis → {funnel_analysis.count():,} rows")
funnel_analysis.show(10)
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# 6. monthly_summary
# ─────────────────────────────────────────────────────────────────────────────
monthly_summary = (
    silver_orders
    .filter(col("is_delivered") == True)
    .groupBy("order_year", "order_month", "order_quarter")
    .agg(
        count("order_id").alias("total_orders"),
        round(sum("total_amount"), 2).alias("total_revenue"),
        round(avg("total_amount"), 2).alias("avg_order_value"),
        sum("quantity").alias("total_items_sold"),
        countDistinct("customer_id").alias("unique_customers"),
        countDistinct("product_id").alias("unique_products_sold"),
        countDistinct("category").alias("active_categories"),
        round(sum(when(col("is_weekend") == True,
                       col("total_amount"))), 2).alias("weekend_revenue"),
        round(sum(when(col("is_weekend") == False,
                       col("total_amount"))), 2).alias("weekday_revenue"),
    )
    .withColumn("prev_month_revenue",
        lag("total_revenue", 1).over(
            Window.partitionBy()
            .orderBy("order_year", "order_month")
        ))
    .withColumn("mom_growth_pct",
        when(col("prev_month_revenue").isNotNull(),
             round(
                 (col("total_revenue") - col("prev_month_revenue"))
                 / col("prev_month_revenue") * 100
             , 2))
        .otherwise(lit(None)))
    .withColumn("prev_year_revenue",
        lag("total_revenue", 12).over(
            Window.partitionBy()
            .orderBy("order_year", "order_month")
        ))
    .withColumn("yoy_growth_pct",
        when(col("prev_year_revenue").isNotNull(),
             round(
                 (col("total_revenue") - col("prev_year_revenue"))
                 / col("prev_year_revenue") * 100
             , 2))
        .otherwise(lit(None)))
    .withColumn("cumulative_revenue",
        round(sum("total_revenue").over(
            Window.partitionBy()
            .orderBy("order_year", "order_month")
            .rowsBetween(Window.unboundedPreceding, Window.currentRow)
        ), 2))
    .withColumn("weekend_revenue_pct",
        round(col("weekend_revenue") / col("total_revenue") * 100, 2))
    .withColumn("silver_updated_at", current_timestamp())
    .orderBy("order_year", "order_month")
)

(
    monthly_summary.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .option("path", "s3://ecommerce-lakehouse-databricks/gold/monthly_summary/")
    .saveAsTable("ecommerce.gold.monthly_summary")
)

print(f"✅ gold.monthly_summary → {monthly_summary.count():,} rows")
monthly_summary.show(10)
```
